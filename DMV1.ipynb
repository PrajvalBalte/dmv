```python
import pandas as pd

df = pd.read_csv("dataset/sales_data_sample.csv", encoding="latin-1")
df.head()
```

```python
df.drop_duplicates()
```

```python
df=df.dropna(subset=["PRODUCTLINE"])
df.head(5)
```

```python
data_pie=df['PRODUCTLINE'].value_counts().rename_axis('PRODUCTLINE').reset_index(name='orders')
```

```python

import matplotlib.pyplot as plt

plt.figure(figsize=(10,10))
plt.pie(data_pie.orders,labels=data_pie.PRODUCTLINE,startangle=90, autopct='%.1f%%')
plt.show()
```

```python
sum1=df['SALES'].sum()
max1=df['SALES'].max()
min1=df['SALES'].min()
print("sum",sum1)
print("max",max1)
print("min",min1)
```

```python
import seaborn as sns
```

```python
import json
from IPython.display import display, JSON

json_file='dataset/sales-data.json'
with open(json_file,'r') as f:
    data=json.load(f)
display(data)
```

```python
newdf=pd.json_normalize(data)
```

```python
print(newdf.columns)
```

```python
df=pd.DataFrame(data['sales'])
```

```python

plt.figure(figsize=(10,6))
plt.scatter(df['quantity'], df['revenue'])
plt.xlabel('Quantity')
plt.ylabel('Revenue')
plt.title('Scatter Plot of Quantity vs Revenue')
plt.show()
```

```python
plt.hist(df['quantity'], bins=5, edgecolor='black')
plt.title('Histogram of Quantity')
plt.xlabel('Quantity')
plt.ylabel('Frequency')
plt.show()
```
